# Resampling 

In timeseries, resampling is the processes of changing the interval of time.  We cover the two main types of resampling.  

## Define downsampling 

> Resample a dataset to a wider timeframe (hours to days to months, etc.)
>
> Use an aggregate function with the resampling
>    - Sum, avg, min, max, etc.

## Define upsampling

> Resampling to smaller timeframe
>
> Do not use aggregate function; would need to fill missing values (back-, forward-fills, interpolate)

## If wanted to resample the gun_crime data to look at weekly counts of gun crime, would you use upsampling or downsampling?

> Downsampling

# Time series modeling requires that our data is stationary.  

## What is stationary data?

> A stationary process has the property that the mean, variance and autocorrelation structure do not change over time. Stationarity can be defined in precise mathematical terms, but for our purpose we mean a flat looking series, without trend, constant variance over time, a constant autocorrelation structure over time and no periodic fluctuations (seasonality). [source](https://www.itl.nist.gov/div898/handbook/pmc/section4/pmc442.htm#:~:text=Stationarity-,Stationarity,-A%20common%20assumption)

## Why does a time series model need stationary data?

> Assume independence of each point - no trends, seasonality
>
> "...Stationarity means that the statistical properties of a process generating a time series do not change over time." [source](https://towardsdatascience.com/stationarity-in-time-series-analysis-90c94f27322#:~:text=In%20the%20most%20intuitive%20sense%2C%20stationarity%20means%20that%20the%20statistical%20properties%20of%20a%20process%20generating%20a%20time%20series%20do%20not%20change%20over%20time)

> [Source](https://www.itl.nist.gov/div898/handbook/pmc/section4/pmc442.htm#:~:text=Transformations%20to%20Achieve%20Stationarity)
>
> 1. We can **difference the data.** That is, given the series Zt, we create the new series:
$Yi=Ziâˆ’Ziâˆ’1.$
>
>  The differenced data will contain one less point than the original data. Although you can difference the data more than once, one difference is usually sufficient.
>
>
> 2. If the data contain a trend, we can **fit some type of curve to the data and then model the residuals from that fit**. 
>    - Since the purpose of the fit is to simply remove long term trend, a simple fit, such as a straight line, is typically used.
>
>
> 3. For non-constant variance, **taking the logarithm or square root of the series may stabilize the variance.**
>    - For negative data, you can add a suitable constant to make all the data positive before applying the transformation.
>        - This constant can then be subtracted from the model to obtain predicted (i.e., the fitted) values and forecasts for future points

# Function from James's SG

```python
## Lab Function
from statsmodels.tsa.stattools import adfuller

def adfuller_test_df(ts,index=['AD Fuller Results']):
    """Returns the AD Fuller Test Results and p-values for the null hypothesis
    that there the data is non-stationary (that there is a unit root in the data)"""
    
    df_res = tsa.stattools.adfuller(ts)

    names = ['Test Statistic','p-value','#Lags Used','# of Observations Used']
    res  = dict(zip(names,df_res[:4]))
    
    res['p<.05'] = res['p-value']<.05
    res['Stationary?'] = res['p<.05']
    
    if isinstance(index,str):
        index = [index]
    res_df = pd.DataFrame(res,index=index)
    res_df = res_df[['Test Statistic','#Lags Used',
                     '# of Observations Used','p-value','p<.05',
                    'Stationary?']]
    return res_df



def stationarity_check(TS,window=8,plot=True,index=['AD Fuller Results']):
    """Adapted from https://github.com/learn-co-curriculum/dsc-removing-trends-lab/tree/solution"""
    
    # Calculate rolling statistics
    roll_mean = TS.rolling(window=window, center=False).mean()
    roll_std = TS.rolling(window=window, center=False).std()
    
    # Perform the Dickey Fuller Test
    dftest = adfuller_test_df(TS,index=index)
    
    if plot:
        # Plot rolling statistics:
        fig = plt.figure(figsize=(12,6))
        plt.plot(TS, color='blue',label=f'Original (freq={TS.index.freq})')
        plt.plot(roll_mean, color='red', label=f'Rolling Mean (window={window})')
        plt.plot(roll_std, color='black', label = f'Rolling Std (window={window})')
        plt.legend(loc='best')
        plt.title('Rolling Mean & Standard Deviation')
        display(dftest)
        plt.show(block=False)
        
    return dftest
    
```

#### Trend Removal Methods
- Differencing (`.diff()`)
- Log-Transformation (`np.log`)
- Subtract Rolling Mean (`ts-ts.rolling().mean()`)
- Subtract Exponentially-Weighted Mean (`ts-ts.ewm().mean()`)
- Seasonal Decomposition (`from statsmodels.tsa.seasonal import seasonal_decompose`
)
